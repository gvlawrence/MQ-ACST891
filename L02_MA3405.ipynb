### Separating source and heritage into different columns
**Step 1.** Select **Source_Heritage menu > Edit column > Add column based on this
column...**<br>

**Step 2.** Set **New column name** to Heritage Listed .<br>
**Step 3.** Set **Expression** to:<br>
*if(value == "Yes", "Yes", "No")*<br>

OpenRefine uses GREL (Google Refine Expression Language). More info on GREL syntax can found here
http://code.google.com/p/google-refine/wiki/UnderstandingExpressions.<br>

The GREL expression used in step 3 is being used to add values to a new column, and these values are based on the
existing values found in the Source_Heritage column. This column is messy and contains various values of mixed
meanings (i.e. the **Source of information in the row:** “LGOV (Local Government)”, “SGOV (State Government”),
“GAZ (NSW Government Gazette)”; and **if the Police Station is Heritage listed:** “Yes”. Here we’re going to start
cleaning the data by separating the mixed meaning of the Source_Heritage column into 2 new columns with specific
meanings: the Source of the data and the Heritage Status of each Police Station.<br>

The first GREL expression is related to the Heritage Status of each Police Station:<br>

**GREL Expression:** <span style="color:red">if</span> ( <span style="color:red">value</span> <span style="color:green">==</span> <span style="color:red">"Yes"</span> , <span style="color:orange">"Yes"</span> , <span style="color:deeppink">"No"</span> )<br>

Meaning: for every row, <span style="color:red">if the value of the cell in column 5</span> <span style="color:green"> equals</span> <span style="color:red">“Yes”</span> <span style="color:blue">, then in the new column add a
value of</span> <span style="color:orange">“Yes”</span> , otherwise in the new column add a value of <span style="color:deeppink">“No”.</span>

![](https://jamescookuniversity-my.sharepoint.com/:i:/r/personal/neil_fraser_jcu_edu_au/Documents/DBMA3405/Open%20Refine/Images/Grel%208.PNG?csf=1&e=OO65e7) <br>
**Step 4.** Select **Source_Heritage menu > Edit column > Add column based on this
column...**<br>
**Step 5.** Set **New column name** to *Source* .<br>
**Step 6.** Set **Expression** to:<br>
*if(value != "Yes", value, "")*<br>

The second GREL expression is to denote the source of information (“LGOV”, “SGOV” or “GAZ”)<br>

**GREL Expression:** <span style="color:red">if</span> ( <span style="color:red">value</span> <span style="color:green">!=</span> <span style="color:deeppink">"Yes"</span> , <span style="color:yellowgreen">value</span> ,  <span style="color:blue">""</span> ) <br>

Meaning: for every row , <span style="color:red"> if the value of the cell in Source_Heritage</span> ) <span style="color:green"> does not equal</span> ) <span style="color:deeppink">  “Yes”</span> ) <span style="color:green"> then in the new column,</span> <span style="color:yellowgreen"> insert a value in the new column of</span> <span style="color:blue">""  (i.e. whatever the value was that appeared in
Source_Heritage)</span> <br>

### Collapsing duplicate rows into a single record
In this and the following section we will work through a process to identify as many duplicate entries from our dataset as we can. Consider the following two entries:  

![grel10](https://user-images.githubusercontent.com/6886376/43748228-c7b7c992-9a31-11e8-9555-278dfab12274.PNG)<br>

Although the station name is different, we can be confident that they are duplicates because they refer to the same address, suburb and LGA. However we can also see that the LGA is slightly different, with an extra ‘o’ in the first entry. OpenRefine’s clustering ability is perfect for dealing with these inconsistencies, but it does not allow us to use clustering across
multiple columns at the same time.<br>

In order to identify and cleanup our duplicates entries we need to create another **compound** column that is a combination of the *Address , Suburb* and *LGA* columns. Using a combination of three locations fields allows us to make sure that, for example, 16
Jersey Road in Paddington is not accidentally merged with 16 Jersey Road in another suburb.
 

The steps to do this are:
**Step 1.** Create the compound column - for example *“16 Jersey Road | Paddington |Woollahrah”*<br>
**Step 2.** Use clustering on this column to identify and merge similar entries - merging *“16 Jersey Road | Paddington | Woollahrah”* with *“16 Jersey Road | Paddington | Wollahrah”*<br>
**Step 3.** Use a feature of OpenRefine called **Records** to identify entries with exactly the same value in this field.<br>

Let’s get started.


### Creating a composite column
**Step 1.** Select **Address menu > Edit Column > Add column based on this column…**<br>
**Step 2.** Set **New column name to** Compound Key<br>
**Step 3.** Set **Expression** to:<br>
value + " | " + cells["Suburb"].value + " | " + cells["LGA"].value

The second GREL expression is to combine the values across three columns (Address, Suburb and LGA)<br>
**GREL Expression:** <span style="color:red">value</span> <span style="color:green">+ “ | ” +</span> <span style="color:deeppink">"cells[“Suburb”].value</span> <span style="color:green">+ “ | ” +</span> <span style="color:yellowgreen">cells[“LGA”].value</span> 

**Meaning:** for every row , <span style="color:red">get the value of the cell in the “Address” column</span> <span style="color:green">add spaces and a “pipe” symbol to this to clearly separate this from the next field,</span> <span style="color:deeppink">add the value of the cell in the Suburb column</span> <span style="color:green">add
more spaces and another pipe,</span>   <span style="color:yellowgreen">then add the value of the cell in the LGA column</span>  

![](https://jamescookuniversity-my.sharepoint.com/:i:/r/personal/neil_fraser_jcu_edu_au/Documents/DBMA3405/Open%20Refine/Images/Grel%206.PNG?csf=1&e=qNn28w)

**Step 4.** Click **OK** to create the column.<br>

### Supplementing the data by calling into an API
It's relatively straightforward to draw in data from an external API with OpenRefine. In this case we'll call in to Google's Geolocation API to get the longitude and latitude of all our police stations.<br>
**Step 1.** In **Rows** view, ensure all rows are displayed by selecting **Reset All** .<br>
**Step 2.** Select **Suburb menu > Edit column > Add column by fetching URLs.**<br>
**Step 3.** Type *Geocoding Response* into **New column name.**<br>
**Step 4.** Type *50* in **Throttle delay.**<br>
**Step 5.** Type the following exactly into the **Expression** box:<br>

"https://maps.googleapis.com/maps/api/geocode/json?sensor=false&address="+escape(value + ", New South Wales", "url")

The language used to query the Google Geocoding Application Programmer Interface (API) is found here:
https://developers.google.com/maps/documentation/geocoding/ .<BR>
    
These state that the format of the request to this API must be:
http://maps.googleapis.com/maps/api/geocode/output?parameters and the minimum parameters required are “address” ( the address that you want to get geo-co-ordinates for ) and “sensor” ( which indicates whether or not the geocoding request comes from a device with a location sensor. This value must be either true or false) . Parameters are separated by the ampersand (&) symbol.

Our Expression:
<span style="color:red">https://maps.googleapis.com/maps/api/geocode/</span> <span style="color:blue">json</span> <span style="color:yellowgreen">?</span> <span style="color:purple">sensor=false</span> <span style="color:peru">&</span> <span style="color:deeppink">address="</span>  +escape <span style="color:green">(value + " </span> <span style="color:orange">(, New South Wales", </span> "url" ) <br>

Meaning:
This is telling your computer to visit <span style="color:red">https://maps.googleapis.com/maps/api/geocode/</span>  and <span style="color:blue">get output in JSON (JavaScript Object Notation) language ,</span> <span style="color:yellowgreen">from a query to the API.</span> <span style="color:purple">We are telling it that the API query request does not come from a device where there is a location sensor (it’s just coming from a list of text terms),</span> <span style="color:peru">and</span> that <span style="color:deeppink">the query address to use</span> <span style="color:green">is the value found in the suburb column,</span> <span style="color:orange">and limit the results to those that only include “New South Wales” somewhere in the information compiled by Google.</span> <br>
    
The escape and url refer to GREL string functions where we can basically invoke a function and then ‘escape out’ in
url mode. See https://github.com/OpenRefine/OpenRefine/wiki/GREL-String-Functions#escapestring-s-string-mode .

![](https://jamescookuniversity-my.sharepoint.com/:i:/r/personal/neil_fraser_jcu_edu_au/Documents/DBMA3405/Open%20Refine/Images/Grel%203.PNG?csf=1&e=mkPRve)<br>

**Step 6.** Click **OK** . OpenRefine will query the API for each row in the dataset.<br>

As above, now we’re just filtering out the Longitude Value Google has on each place:<br>


**Expression:** <span style="color:blue">parseJson(</span> <span style="color:orange">value</span> <span style="color:blue">)</span> .results[0]. <span style="color:green">geometry</span> <span style="color:red">.location.</span> <span style="color:orange">lng</span>

**Meaning:** <span style="color:blue">Parse the JSON results in each cell of the table</span> to only show the <span style="color:orange">longitude (lng) value</span> , 
which is nested under <span style="color:red">the location information</span> , which is nested under <span style="color:green">the geometry information</span> . Again, we’re only interested in getting the data from the first ([0]) result that was retrieved.

### Undo/Redo History
OpenRefine has an infinite undo history. To access:
**Step 1.** Click on the **Undo/Redo** tab. You'll see every action you've done since creating the project.<br>
**Step 2.** You can undo to any step by clicking on the step you want to revert back to. Similarly, you can redo every step.

If you wish to save a set of steps to be re-applied later, or to a different project, you can click the **Extract** button. This gives you the option to select the steps you want to save, and to copy the transformations included in the selected steps in a format called ‘JSON’<br>

To apply a set of steps you have copied or saved in this ‘JSON’ format use the **Apply** button and paste in the JSON. In this way you can share transformations between projects and each other.

### Exporting the dataset
**Step 1.** Click **Export** in the top right-hand corner.<br>
**Step 2.** Select **Comma-separated variable** from the drop-down menu. A CSV dump of your data will be downloaded.<br>
